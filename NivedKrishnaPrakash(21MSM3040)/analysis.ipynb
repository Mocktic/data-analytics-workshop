```python
import pandas as pd
import matplotlib as mpl
import numpy as np 
from matplotlib import pyplot as plt
import statistics
plt.rcParams["figure.figsize"] = [20, 8]
```

```python
data= pd.read_csv('Data/Combined_Clean.csv')
```

```python
df=pd.DataFrame(data)
```

```python
df.shape
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna(inplace= True)
df.reset_index(drop=True)
```

```python
df["Acre Value"].isna().values.any()
```

```python
print("Mean: ",np.mean(df['Acre Value']))
print("Median: ",statistics.median(df['Acre Value']))
print("Mode: ",statistics.mode(df['Acre Value']))
print("Standard Deviation: ",statistics.stdev(df['Acre Value']))
```

```python
year_df=pd.DataFrame(df.groupby(['Year']).sum())
year_df['Acre Value'].plot()
plt.xlabel('Year')
plt.ylabel('Acre Value')
```

```python
plt.bar(df['LandCategory'],df['Acre Value'])
plt.title('Acre Value vs LandCategory')
plt.xlabel('LandCategory')
plt.ylabel('Acre Value')
```

```python
plt.xticks(rotation=45, ha='right')
plt.bar(df['Region'],df['Acre Value'])
plt.title('Acre Value vs Region')
plt.xlabel('Region')
plt.ylabel('Acre Value')
```

```python
plt.xticks(rotation=45, ha='right')
plt.bar(df['State'],df['Acre Value'])
plt.xlabel('State')
plt.ylabel('Acre Value')
```

```python
plt.boxplot(year_df['Acre Value'])
plt.ylabel('Acre Value')
```
